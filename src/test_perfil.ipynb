````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
````

### 1. Equação 1

\[
E_l(f) = 0.125 C_T^2 f_l^{-5/3}
\]

**Explicação**:
- Esta equação descreve a densidade espectral de energia das flutuações de temperatura do ar em função da frequência \(f\).
- **\(C_T\)** é o parâmetro de estrutura das flutuações de temperatura do ar.
- **\(f_l\)** é a escala espacial na faixa espectral de pequenas escalas.
- Utilizada para modelar a turbulência atmosférica em pequena escala baseada na teoria de Kolmogorov.

#### 2. Equação 2

\[
E_l(f) = E(f_L) \exp\left(-3 \ln^2 \left(\frac{f_l}{f_t}\right)\right)
\]

**Explicação**:
- Modela a densidade espectral de energia para a faixa de alta frequência.
- **\(f_L\)** é a frequência da faixa de baixa frequência espacial.
- **\(f_t\)** é a frequência na região espectral de transição.
- Ajuda a entender a transição entre diferentes regimes de turbulência.

#### 3. Equação 3

\[
f_T = 0.7 V f_L
\]

**Explicação**:
- Relaciona a frequência temporal de referência **\(f_T\)** com a frequência espacial **\(f_L\)** e a velocidade média do vento **\(V\)**.
- Utilizada para converter dados de frequência espacial para frequência temporal.

#### 4. Equação 4

\[
\sigma_T^2 = \int_{f_T}^{f_m} E(f) df = \int_{f_T}^{f_m} E(f) df
\]

**Explicação**:
- Calcula a variância da temperatura do ar **\(\sigma_T^2\)** a partir da densidade espectral de energia.
- **\(f_T\)** e **\(f_m\)** são as frequências de referência para as integrais.
- Importante para estimar a energia de flutuações de temperatura em diferentes escalas.

#### 5. Equação 5

\[
E_l(f) = 0.125 C_T^2 f_l^{-5/3} \exp\left(-3 \ln^2 \left(\frac{f_l}{f_t}\right)\right)
\]

**Explicação**:
- Combina os aspectos das Equações 1 e 2 para fornecer um modelo mais abrangente da densidade espectral de energia.
- Utilizada para perfilar verticalmente o parâmetro de estrutura das flutuações do índice de refração do ar.

#### 6. Equação 6

\[
\sigma_T^2(f_L) = a(f_L, V) \sigma_T^2
\]

**Explicação**:
- Relaciona a variância da temperatura do ar em uma banda de frequência estreita **\(\sigma_T^2(f_L)\)** com a variância total **\(\sigma_T^2\)** e um coeficiente de proporcionalidade **\(a(f_L, V)\)**.
- **\(a(f_L, V)\)** depende da escala da mudança de inclinação do espectro e da velocidade média do vento.
- Utilizada para calcular a densidade espectral de flutuações em uma banda de frequência estreita a partir da variância de temperatura.

### Utilização das Equações

Essas equações são utilizadas para modelar a estrutura da turbulência atmosférica, essencial para sistemas de óptica adaptativa. Eles ajudam a prever como a turbulência afeta a propagação de ondas ópticas e a ajustar sistemas ópticos para minimizar essas distorções, melhorando a qualidade das observações astronômicas e outras aplicações relacionadas.

````python


# Definindo variáveis
f, f_L, f_t, V, sigma_T, C_T = sp.symbols('f f_L f_t V sigma_T C_T')
E = sp.Function('E')(f)
E_L = sp.Function('E_L')(f_L)
a = sp.Function('a')(f_L, V)

# Equação 1
equation_1 = sp.Eq(E, 0.125 * C_T**2 * f**(-5/3))
sp.pretty_print(equation_1)

# Equação 2
equation_2 = sp.Eq(E_L, E * sp.exp(-3 * sp.ln(f / f_t)**2))
sp.pretty_print(equation_2)

# Equação 3
f_T = sp.symbols('f_T')
equation_3 = sp.Eq(f_T, 0.7 * V * f_L)
sp.pretty_print(equation_3)

# Equação 4
f_m = sp.symbols('f_m')
equation_4 = sp.Eq(sigma_T**2, sp.Integral(E, (f, f_T, f_m)))
sp.pretty_print(equation_4)

# Equação 5
equation_5 = sp.Eq(E, 0.125 * C_T**2 * f**(-5/3) * sp.exp(-3 * sp.ln(f / f_t)**2))
````

Para abordar este problema, primero vamos a definir las ecuaciones y luego a reemplazar los valores proporcionados en ellas usando Python. También generaremos gráficos para visualizar los resultados.

### Ecuaciones
1. **Ecuación 1**
\[ E_l(f_{ls}) = 0.125 C_T^2 f_{ls}^{-5/3} \]

2. **Ecuación 2**
\[ E_l(f_l) = E(f_L) \exp\left( -\frac{3}{5} \ln^2 \left( \frac{f_t}{f_L} \right) \right) \]

3. **Ecuación 3**
\[ f_L = 0.7 f_T V \]

4. **Ecuación 4**
\[ \sigma_T^2(f_L) = E(f_L) f_L \]

5. **Ecuación 5**
\[ E_L(f) = a \sigma_T^2(f) \]

6. **Ecuación 6**
\[ C_n^2(z) = C_{n0}^2 \exp\left( -\frac{z}{h_0} \right) \]

7. **Ecuación 7**
\[ H_{\text{eff}} = \left( \frac{\int_0^{H} C_n^2(z) z^5 dz}{\int_0^{H} C_n^2(z) dz} \right)^{1/5} \]

### Valores proporcionados
- **Temperatura (T)**: 500 (constante)
- **Velocidad del viento (V)**: 100 a 5000
- **Presión (P)**: 30 a 530

### Código en Python


````python



# Definir variables simbólicas
C_T, f_ls, E_l, E_fL, f_t, f_L, f_T, V, sigma_T, a, C_n0, z, h_0, H = sp.symbols('C_T f_ls E_l E_fL f_t f_L f_T V sigma_T a C_n0 z h_0 H')

# Definir constantes
C_T_value = 500  # Temperatura
f_t_value = 1  # Frecuencia transitoria arbitraria para simplificación
C_n0_value = 1  # Valor arbitrario para C_n0
h_0_value = 1  # Valor arbitrario para h_0

# Rango de valores para la presión y la velocidad del viento
velocidades_viento = np.linspace(100, 5000, 100)
presiones = np.linspace(30, 530, 100)

# Ecuación 1
E_l_1 = 0.125 * C_T_value**2 * presiones**(-5/3)

# Ecuación 2
E_l_2 = E_fL * np.exp(-3/5 * np.log(f_t_value / presiones)**2)

# Ecuación 3
f_L_3 = 0.7 * f_T * velocidades_viento

# Ecuación 4
sigma_T_4 = E_fL * f_L

# Ecuación 5
E_l_5 = a * sigma_T**2

# Ecuación 6
C_n_6 = C_n0_value**2 * np.exp(-velocidades_viento / h_0_value)

# Ecuación 7
H_eff_7 = (sp.integrate(C_n0_value**2 * sp.exp(-z / h_0_value) * z**5, 
                        (z, 0, H)) / sp.integrate(C_n0_value**2 * sp.exp(-z / h_0_value), (z, 0, H)))**(1/5)

# Calcular resultados para ecuaciones 2, 3, 4, 5 y 7
E_fL_value = 1  # Valor arbitrario para E_fL
a_value = 1  # Valor arbitrario para a

# Ecuación 2 con valores
E_l_2_values = E_fL_value * np.exp(-3/5 * np.log(f_t_value / presiones)**2)

# Ecuación 3 con valores
f_L_3_values = 0.7 * C_T_value * velocidades_viento

# Ecuación 4 con valores
sigma_T_4_values = E_fL_value * f_L_3_values

# Ecuación 5 con valores
E_l_5_values = a_value * sigma_T_4_values**2

# Ecuación 7 con valores
H_value = 1000  # Altura arbitraria para simplificación
H_eff_7_value = (sp.integrate(C_n0_value**2 * sp.exp(-z / h_0_value) * z**5, 
                              (z, 0, H_value)) / sp.integrate(C_n0_value**2 * sp.exp(-z / h_0_value), (z, 0, H_value)))**(1/5)

# Convertir H_eff_7 a valor numérico
H_eff_7_value_num = H_eff_7_value.evalf()

# Imprimir resultados
print(f"Ecuación 1: {E_l_1}")
print(f"Ecuación 2: {E_l_2_values}")
print(f"Ecuación 3: {f_L_3_values}")
print(f"Ecuación 4: {sigma_T_4_values}")
print(f"Ecuación 5: {E_l_5_values}")
print(f"Ecuación 6: {C_n_6}")
print(f"Ecuación 7: {H_eff_7_value_num}")

# Graficar resultados
plt.figure(figsize=(12, 8))

# Gráfico para Ecuación 1
plt.subplot(2, 3, 1)
plt.plot(presiones, E_l_1)
plt.title('Ecuación 1')
plt.xlabel('Presión')
plt.ylabel('E_l')

# Gráfico para Ecuación 2
plt.subplot(2, 3, 2)
plt.plot(presiones, E_l_2_values)
plt.title('Ecuación 2')
plt.xlabel('Presión')
plt.ylabel('E_l')

# Gráfico para Ecuación 3
plt.subplot(2, 3, 3)
plt.plot(velocidades_viento, f_L_3_values)
plt.title('Ecuación 3')
plt.xlabel('Velocidad del viento')
plt.ylabel('f_L')

# Gráfico para Ecuación 4
plt.subplot(2, 3, 4)
plt.plot(velocidades_viento, sigma_T_4_values)
plt.title('Ecuación 4')
plt.xlabel('Velocidad del viento')
plt.ylabel('sigma_T^2')

# Gráfico para Ecuación 5
plt.subplot(2, 3, 5)
plt.plot(velocidades_viento, E_l_5_values)
plt.title('Ecuación 5')
plt.xlabel('Velocidad del viento')
plt.ylabel('E_l')

# Gráfico para Ecuación 6
plt.subplot(2, 3, 6)
plt.plot(velocidades_viento, C_n_6)
plt.title('Ecuación 6')
plt.xlabel('Velocidad del viento')
plt.ylabel('C_n^2')

plt.tight_layout()
plt.show()
````



### Explicación de las Ecuaciones y Valores
- **Temperatura (T)**: Se usa en la ecuación 1 como parte de \( C_T \) y en la ecuación 3 como parte de \( f_T \).
- **Velocidad del viento (V)**: Se usa en la ecuación 3 para calcular \( f_L \) y en la ecuación 6 para calcular \( C_n^2(z) \).
- **Presión (P)**: Se usa en la ecuación 1 para calcular \( E_l(f_{ls}) \) y en la ecuación 2 para calcular \( E_l(f_l) \).

Este código permite calcular y graficar los valores de las ecuaciones 1 a 6, usando los valores proporcionados para temperatura,
velocidad del viento y presión. La ecuación 7 también se calcula y se muestra su valor numérico.

````python


# Definir variables simbólicas
C_T, f_ls, E_l, E_fL, f_t, f_L, f_T, V, sigma_T, a, C_n0, z, h_0, H = sp.symbols('C_T f_ls E_l E_fL f_t f_L f_T V sigma_T a C_n0 z h_0 H')

# Ecuación 1
equation_1 = E_l - 0.125 * C_T**2 * f_ls**(-5/3)

# Ecuación 2
equation_2 = E_l - E_fL * sp.exp(-3/5 * sp.ln(f_t / f_L)**2)

# Ecuación 3
equation_3 = f_L - 0.7 * f_T * V

# Ecuación 4
equation_4 = sigma_T**2 - E_fL * f_L

# Ecuación 5
equation_5 = E_l - a * sigma_T**2

# Ecuación 6
equation_6 = C_n0**2 * sp.exp(-z / h_0)

# Ecuación 7
integral_1 = sp.integrate(equation_6 * z**5, (z, 0, H))
integral_2 = sp.integrate(equation_6, (z, 0, H))
H_eff = (integral_1 / integral_2)**(1/5)

# Valores proporcionados
temperatura = 500
velocidades_viento = np.linspace(100, 5000, 100)
presion = np.linspace(30, 530, 100)

# Calcular resultados para diferentes velocidades del viento y presiones
resultados_1 = []
resultados_2 = []
resultados_3 = []
resultados_4 = []
resultados_5 = []
resultados_6 = []
resultados_7 = []

for v in velocidades_viento:
    for p in presion:
        # Sustitución en las ecuaciones
        res_1 = equation_1.subs({C_T: temperatura, f_ls: p})
        res_2 = equation_2.subs({E_fL: temperatura, f_t: p, f_L: v})
        res_3 = equation_3.subs({f_T: temperatura, V: v})
        res_4 = equation_4.subs({E_fL: temperatura, f_L: v})
        res_5 = equation_5.subs({a: p, sigma_T: temperatura})
        res_6 = equation_6.subs({C_n0: temperatura, z: v, h_0: p})
        res_7 = H_eff.subs({C_n0: temperatura, h_0: p, H: v})
        
        resultados_1.append(res_1.evalf())
        resultados_2.append(res_2.evalf())
        resultados_3.append(res_3.evalf())
        resultados_4.append(res_4.evalf())
        resultados_5.append(res_5.evalf())
        resultados_6.append(res_6.evalf())
        resultados_7.append(res_7.evalf())

# Graficar los resultados
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(velocidades_viento, resultados_1[:100], label='Ecuación 1')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(velocidades_viento, resultados_2[:100], label='Ecuación 2')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(velocidades_viento, resultados_3[:100], label='Ecuación 3')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(velocidades_viento, resultados_4[:100], label='Ecuación 4')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 4')
plt.legend()

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(velocidades_viento, resultados_5[:100], label='Ecuación 5')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 5')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(velocidades_viento, resultados_6[:100], label='Ecuación 6')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 6')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(velocidades_viento, resultados_7[:100], label='Ecuación 7')
plt.xlabel('Velocidad del Viento')
plt.ylabel('Resultado')
plt.title('Ecuación 7')
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definir variables simbólicas
C_T, f_ls, E_l, E_fL, f_t, f_L, f_T, V, sigma_T, a, C_n0, z, h_0, H = sp.symbols('C_T f_ls E_l E_fL f_t f_L f_T V sigma_T a C_n0 z h_0 H')

# Ecuación 1
equation_1 = E_l - 0.125 * C_T**2 * f_ls**(-5/3)

# Ecuación 2
equation_2 = E_l - E_fL * sp.exp(-3/5 * sp.ln(f_t / f_L)**2)

# Ecuación 3
equation_3 = f_L - 0.7 * f_T * V

# Ecuación 4
equation_4 = sigma_T**2 - E_fL * f_L

# Ecuación 5
equation_5 = E_l - a * sigma_T**2

# Ecuación 6
equation_6 = C_n0**2 * sp.exp(-z / h_0)

# Ecuación 7
integral_1 = sp.integrate(equation_6 * z**5, (z, 0, H))
integral_2 = sp.integrate(equation_6, (z, 0, H))
H_eff = (integral_1 / integral_2)**(1/5)

# Sustituir valores
temperature = 500
wind_speeds = np.linspace(100, 5000, 100)
pressures = np.linspace(30, 530, 100)

# Convertir a función numérica
equation_1_num = sp.lambdify((C_T, f_ls), equation_1.rhs)
equation_2_num = sp.lambdify((E_fL, f_t, f_L), equation_2.rhs)
equation_3_num = sp.lambdify((f_T, V), equation_3.rhs)
equation_4_num = sp.lambdify((E_fL, f_L), equation_4.rhs)
equation_5_num = sp.lambdify((a, sigma_T), equation_5.rhs)
equation_6_num = sp.lambdify((C_n0, z, h_0), equation_6)
H_eff_num = sp.lambdify((C_n0, h_0, H), H_eff)

# Calcular resultados
results_1 = equation_1_num(temperature, pressures)
results_2 = equation_2_num(temperature, pressures, wind_speeds)
results_3 = equation_3_num(temperature, wind_speeds)
results_4 = equation_4_num(temperature, wind_speeds)
results_5 = equation_5_num(pressures, temperature)
results_6 = equation_6_num(temperature, wind_speeds, pressures)
results_7 = H_eff_num(temperature, pressures, wind_speeds)

# Graficar resultados
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(pressures, results_1, label='Ecuación 1')
plt.xlabel('Presión (f_ls)')
plt.ylabel('E_l')
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(pressures, results_2, label='Ecuación 2')
plt.xlabel('Presión (f_t)')
plt.ylabel('E_l')
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(wind_speeds, results_3, label='Ecuación 3')
plt.xlabel('Velocidad del Viento (V)')
plt.ylabel('f_L')
plt.legend()

plt.subplot(2, 3, 4)
plt.plot(wind_speeds, results_4, label='Ecuación 4')
plt.xlabel('Velocidad del Viento (f_L)')
plt.ylabel('σ_T^2')
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(pressures, results_5, label='Ecuación 5')
plt.xlabel('Presión (a)')
plt.ylabel('E_l')
plt.legend()

plt.subplot(2, 3, 6)
plt.plot(wind_speeds, results_7, label='Ecuación 7')
plt.xlabel('Velocidad del Viento (H)')
plt.ylabel('H_eff')
plt.legend()

plt.tight_layout()
plt.show()
````
