```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shapes = gpd.read_file("/home/edith/PROGRAMMING_LEARNING/GSST_daytime_sky_quality/data/geo/BR_Municipios/BR_Municipios_2022.shp")
shapes
```

```python
municipalities_mg = shapes.query("SIGLA_UF == 'MG'")
municipalities_mg.plot()
```

```python
brazopolis = municipalities_mg.query("NM_MUN == 'Brazópolis'")
brazopolis.plot()
```

```python
fig, ax = plt.subplots(figsize=(10, 12))

municipalities_mg.plot(ax=ax, color='white', edgecolor='black')
brazopolis.plot(ax=ax, color='red')
```
