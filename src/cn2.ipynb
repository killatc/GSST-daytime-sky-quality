```python

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

def calcular_E_fl(E_fL, f_t, f_l, f_L):
    """
    Calcula la densidad espectral E(f_l) usando la ecuación (5) del paper.

    Parámetros:
    E_fL (float): Densidad espectral en la frecuencia f_L.
    f_t (float): Frecuencia de transición.
    f_l (float): Frecuencia en el rango de alta frecuencia.
    f_L (float): Frecuencia en el rango de baja frecuencia.

    Retorna:
    float: Densidad espectral E(f_l).
    """
    ln_ratio_ft_fl = -3*( math.log(f_t / f_l))
    ln_ratio_ft_fL = - 5/3*(math.log(f_l / f_t))

    exponent =  ln_ratio_ft_fl * ln_ratio_ft_fL
    E_fl = (E_fL * math.exp(exponent) * f_l**(5/3))/0.125
    return E_fl

# Ejemplo de uso
E_fL = 1.0  # Densidad espectral en f_L (ejemplo)
f_t = 1e-5  # Frecuencia de transición (ejemplo)
f_l = 1e-4  # Frecuencia en el rango de alta frecuencia (ejemplo)
f_L = 1e-6  # Frecuencia en el rango de baja frecuencia (ejemplo)

E_fl = calcular_E_fl(E_fL, f_t, f_l, f_L)
print(f"Densidad espectral E(f_l): {E_fl}")
```

```python


def calcular_E_fl(E_fL, f_t, f_l, f_L):
    """
    Calcula la densidad espectral E(f_l) usando la ecuación (5) del paper.

    Parámetros:
    E_fL (float): Densidad espectral en la frecuencia f_L.
    f_t (float): Frecuencia de transición.
    f_l (float): Frecuencia en el rango de alta frecuencia.
    f_L (float): Frecuencia en el rango de baja frecuencia.

    Retorna:
    float: Densidad espectral E(f_l).
    """
    ln_ratio_ft_fl = -3*( math.log(f_t / f_l))
    ln_ratio_ft_fL = - 5/3*(math.log(f_l / f_t))

    exponent =  ln_ratio_ft_fl * ln_ratio_ft_fL
    E_fl = (E_fL * math.exp(exponent) * f_l**(5/3))/0.125
    return E_fl

# Ejemplo de uso
E_fL = 1.0  # Densidad espectral en f_L (ejemplo)
f_t = 1e-5  # Frecuencia de transición (ejemplo)
f_L = 1e-6  # Frecuencia en el rango de baja frecuencia (ejemplo)

# Crear un rango de valores para f_l en las condiciones especificadas (de −3 a −5/3)
f_l_values = np.linspace(-3, -5/3, 100)

# Calcular E(f_l) para cada valor de f_l
E_fl_values = [calcular_E_fl(E_fL, f_t, 10**f_l, f_L) for f_l in f_l_values]

# Graficar los resultados
plt.plot(f_l_values, E_fl_values)
plt.xlabel('log10(f_l)')
plt.ylabel('E(f_l)')
plt.title('Densidad Espectral E(f_l)')
plt.grid(True)
plt.show()
```
